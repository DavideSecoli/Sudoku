## Part One
You should write all of your code for solving sudokus below this cell.

You must include a function called `sudoku_solver(sudoku)` which takes one sudoku puzzle (a 9x9 NumPy array) as input, and returns the solved sudoku as another 9x9 NumPy array. This is the function which will be tested. 

```python


def comb_elements(X, Y):
    "Comb product of elements in X and elements in Y."
    return [x + y for x in X for y in Y]


digits = '123456789'
rows = 'ABCDEFGHI'
cols = digits
squares = comb_elements(rows, cols)
elements = ([comb_elements(rows, col) for col in cols] + 
            [comb_elements(row, cols) for row in rows] +
            [comb_elements(row_sq, col_sq) for row_sq in ('ABC','DEF','GHI') \
             for col_sq in ('123','456','789')])

unsolved = np.array([[-1 for _ in range(9)] for _ in range(9)])

dict_comp = dict((sq, [unit for unit in elements if sq in unit])
             for sq in squares)

groups = dict((sq, set(sum(dict_comp[sq],[])) - set([sq]))
             for sq in squares)


##############################################################
####################### Parse Numpy  #########################
##############################################################


def parse_board(board):
    """Convert numpy board to a dict of possible board_values, {square: digits}. Return 
    False if an issue is detected """
    
    board = ''.join(map(str, board))
    
    # Every square can be any digit. Assign values_left from the board.
    values_left = dict((sq, digits) for sq in squares)
    for sq, digit in grid_values(board).items():
        if digit in digits and not allocate(values_left, sq, digit):
            return False # Fail if we can't assign digit to square sq
    return values_left

def grid_values(board):
    "Convert board into a dict with '0' or '0.' for empties"
    chars = [c for c in board if c in digits or c in '0.']
    assert len(chars) == 81
    return dict(zip(squares, chars))

##############################################################
################ Apply Constraint Propagation ################
##############################################################


def allocate(values_left, sq, digit):
    """Eliminate all values except for 'digit' from values_left[sq] and propagate.
    Return values_left or False if an issue is detected """
    
    other_values = values_left[sq].replace(digit, '')
    if all(delete(values_left, sq, d2) for d2 in other_values):
        return values_left
    else:
        return False

def delete(values_left, sq, digit):
    """Delete digit from values_left[sq]. Propagate when values_left or places <= 2.
    Return values_left or False if a issue is detected"""
    
    if digit not in values_left[sq]:
        return values_left # Already deleted
    values_left[sq] = values_left[sq].replace(digit,'')
    
    # 1. If a square sq is reduced to one value d2, then delete d2 from 'groups'.
    if len(values_left[sq]) == 0:
        return False # Issue: last value removed
    elif len(values_left[sq]) == 1:
        digit2 = values_left[sq]
        if not all(delete(values_left, sq2, digit2) for sq2 in groups[sq]):
            return False
        
    # 2. If unit is reduced to only one place for a value d, then put it there.
    for unit in dict_comp[sq]:
        digit_places = [sq for sq in unit if digit in values_left[sq]]
        if len(digit_places) == 0:
            return False # Issue: no place for this value
        elif len(digit_places) == 1:
            # assign digit to the only one place in unit
            if not allocate(values_left, digit_places[0], digit):
                return False
    return values_left


##############################################################
########################### Solve ############################
##############################################################

def sudoku_solver(board): 
    return np.array(solve(parse_board(board))) \
if np.array(solve(parse_board(board))).shape == (9,9) \
else unsolved

def solve(values_left):
    "Using depth-first search and propagation, try all possible values"
    if values_left is False:
        return False # Failed before
    if all(len(values_left[sq]) == 1 for sq in squares):
        
        l = [int(x) for x in values_left.values()]
        step = 9
        values_left = ([x for x in [l[start:start+step] for start in range(0,len(l),step)]])
        return values_left # Solution
    
    # Check the unfilled square sq with possibilities left
    n, sq = min((len(values_left[sq]), sq) for sq in squares if len(values_left[sq]) > 1)
    
    return filter_elements(solve(allocate(values_left.copy(), sq, digit)) 
                           for digit in values_left[sq])


def filter_elements(sequence):
    "Return elements in sequence equal to true"
    for element in sequence:
        if element: return element
    return False
```
